```python
import random
from io import StringIO

import matplotlib
from matplotlib import pyplot as plt
import requests
from Bio import Phylo
from ete3 import Tree, TreeStyle, NodeStyle
```

11. Считаем дерево https://www.jasondavies.com/tree-of-life/life.txt;

```python
raw_tree = StringIO(requests.get('https://www.jasondavies.com/tree-of-life/life.txt').text)
tree1 = Phylo.read(raw_tree, "newick")   
```

12. Нарисуем это дерево псевдографикой (draw_ascii);

```python
Phylo.draw_ascii(tree1)
```

13. Нарисуем дерево с помощью draw (картинка должна быть внутри скомпилированного документа);

```python
Phylo.draw(tree1)
```

14. Сохраним изображение дерева в растровый формат (png) и векторный (svg);

```python
fig = plt.figure(figsize=(10, 20), dpi=300)
Phylo.draw(tree1, do_show=False)
plt.savefig("py_tree1_phylo.png")

fig = plt.figure(figsize=(10, 20), dpi=300)
Phylo.draw(tree1, do_show=False)
plt.savefig("py_tree1_phylo.svg")
```

15. Нарисуем дерево в более-менее читаемом виде;

```python
fig = plt.figure(figsize=(30, 30), dpi=600)
matplotlib.rc('font', size=2)
matplotlib.rc('lines', linewidth=0.5)
Phylo.draw(tree1, do_show=False)
```

```python
fig = plt.figure(figsize=(30, 30), dpi=600)
matplotlib.rc('font', size=2)
matplotlib.rc('lines', linewidth=0.5)
Phylo.draw(tree1, do_show=False)
plt.savefig("py_tree1_phylo_visible.svg")
```

16. Считаем простое дерево `(((A,B),(C,D)),E);` из текста;

```python
simpletree = Tree("(((A,B), (C,D)), E);")
```

17. Сохраним это дерево в файл;

```python
simpletree.render("simpletree_ete.png", w=183, units="mm");
```

18. Прочитаем дерево https://www.jasondavies.com/tree-of-life/life.txt и нарисуем это дерево с настройками по умолчанию;

```python
raw_tree = requests.get('https://www.jasondavies.com/tree-of-life/life.txt').text
tree2 = Tree(raw_tree, format=1)
#tree2.show()
tree2.render("tree2.png", w=183, units="mm");
```

19. Нарисуем это дерево круговым;

```python
circular_style = TreeStyle()
circular_style.mode = "c"
circular_style.scale = 20
tree2.render("py_tree2_ete3_circ.png", tree_style=circular_style);
```

20. Нарисуем treeoflife с дополнительным выделением позвоночных.

```python
n1 = tree2.get_common_ancestor("Homo_sapiens", "Danio_rerio")
nst1 = NodeStyle()
nst1["bgcolor"] = "LightSteelBlue"
n1.set_style(nst1)
#tree2.show(tree_style=circular_style)
tree2.render("py_tree2_ete3_vertebrates.png", tree_style=circular_style);
```
